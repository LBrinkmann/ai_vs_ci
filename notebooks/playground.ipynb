```python
import networkx as nx
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
G = nx.random_regular_graph(4, 10)
G = nx.watts_strogatz_graph(10, 4, 0)
nx.draw(G)
plt.draw()
```

```python
c_numbers = []
for i in range(1000):
    G = nx.watts_strogatz_graph(10, 4, 0)
    colors = nx.algorithms.coloring.greedy_color(G)
    c_number = max(colors.values())
    c_numbers.append(c_number)
sns.countplot(c_numbers)
```
